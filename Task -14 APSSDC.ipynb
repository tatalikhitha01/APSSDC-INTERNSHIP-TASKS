```python
import numpy as np
import pandas as pd
```

```python
titanic.isnull().sum()
```

```python
titanic.columns
```

```python
titanic[['Cabin','Embarked']].isnull().sum()
```

```python
titanic['Embarked'].value_counts()
```

```python
titanic['Embarked'] = titanic['Embarked'].fillna(titanic['Embarked'].value_counts().index[0])
```

```python
titanic['Cabin'].value_counts()
```

### Task3
  - Display the plot of original data and robust scaler data

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
df
```

```python
df.drop(['Unnamed: 0'],axis = 1,inplace = True)
df
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rs = RobustScaler()
rs
```

```python
rs1 = rs.fit_transform(df)
rs1
```

```python
rs1 = pd.DataFrame(rs1, columns = df.columns)
rs1
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))
```

```python
fig,(ax1,ax2) = plt.subplots(ncols = 2, figsize = (9,5))

sns.kdeplot(df['TV'], ax = ax1)
sns.kdeplot(df['radio'], ax = ax1)
sns.kdeplot(df['newspaper'], ax = ax1)
sns.kdeplot(df['sales'], ax = ax1)

sns.kdeplot(rs1['TV'], ax = ax2)
sns.kdeplot(rs1['radio'], ax = ax2)
sns.kdeplot(rs1['newspaper'], ax = ax2)
sns.kdeplot(rs1['sales'], ax = ax2)    ## Robust Scaler
```
