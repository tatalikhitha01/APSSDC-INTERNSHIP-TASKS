## Task 
 - Take one dataset and plot all graphs

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Line Plot

```python
data = pd.read_csv("birds.csv")
data
```

```python
plt.plot(data["feml"][:])
plt.title("BIRDS")
plt.show()
```

## Scatter Plot

```python
plt.scatter(data["huml"][:],data["humw"][:])
plt.xlabel('Data of huml',color='red')
plt.ylabel('Data of humw',color='red')
plt.title("Bird Dataset")
plt.show()
```

## Histogram

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
data["type"] = le.fit_transform(data["type"])
data
```

```python
data["type"].value_counts()
```

```python
plt.hist(data["type"], width = 0.6)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

## Bar Graph

```python
plt.bar(data["type"],data["tarw"])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bird Dataset")
plt.show()
```

## Pie Chart

```python
plt.pie(data["huml"][:10],labels = data["ulnal"][:10])
plt.show()
```

```python
plt.pie(data["huml"][:5],labels = data["ulnal"][:5],autopct = "% 1.1f%%")
plt.show()
```

## Box Plot

```python
plt.boxplot(data["type"])
plt.xlabel("X-axis",color='green')
plt.ylabel("Y-axis",color='green')
plt.title("Bird Dataset")
plt.show()
```
