## Apply Random forest model to above dataset.


```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Glass_Classification.csv')
```

```python
X = df.drop('Type', axis = 1) ## Input variable
Y = df['Type']    ## Target variable
```

```python
X.head()
```

```python
Y[:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.30, random_state = 45)
```

```python
X_train
```

```python
X_train.shape , X_test.shape
```

```python
Y_train.shape, Y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc
```

```python
rfc.fit(X_train, Y_train)
```

```python
Y_prediction = rfc.predict(X_test)
Y_prediction
```

```python
X_test.head(3)
```

```python
Y[:3]
```

```python
rfc.predict([[1.51594,13.09,3.52,1.55,72.87,0.68,8.05,0.0,0.09]])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_prediction, Y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_prediction, Y_test)
```
